## Each line of poem (including title) as its own string

```python
# create column using custom function
df['clean_lines'] = df.poem_lines.apply(line_cleaner)
df.clean_lines
```

```python
df['lines_titled'] = df['title'].apply(lambda x: [str(x).upper()+'\n']) + df['clean_lines']
df.lines_titled
```

```python
df['string_titled'] = df.lines_titled.apply(lambda x: '\n'.join(x))
df.string_titled
```

```python
big_string = '\n\n\n\n'.join(df.string_titled)
print(big_string[99500:102500])
```
